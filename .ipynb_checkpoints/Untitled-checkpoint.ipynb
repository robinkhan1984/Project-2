```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
import datetime
import calendar
```

```python
# Load up excel data files 

coal_df = pd.read_excel('Resources/coal.xlsx')
natural_gas_df = pd.read_excel('Resources/Natural_Gas.xlsx')
petroleum_df = pd.read_excel('Resources/Petroleum.xlsx')
renewables_df = pd.read_excel('Resources/Renewable_Energy.xlsx')
```

```python
# Clean out Coal data
clean_coal = coal_df.drop(["Waste Coal Supplied (Thousand Short Tons)","Coal Imports (Thousand Short Tons)","Coal Exports (Thousand Short Tons)","Coal Losses and Unaccounted for (Thousand Short Tons)"],axis=1)
clean_coal
```

```python
# Clean out Natural gas data
clean_natural_gas = natural_gas_df.drop(["Natural Gas Gross Withdrawals (Billion Cubic Feet)","NGPL Production (Billion Cubic Feet)","Natural Gas Production (Dry) (Billion Cubic Feet)","Natural Gas Net Imports (Billion Cubic Feet)", "Supplemental Gaseous Fuels (Billion Cubic Feet)","Natural Gas Balancing Item (Billion Cubic Feet)" ],axis=1)
clean_natural_gas
```

```python
# Clean out Petroleum data
clean_petroleum = petroleum_df.drop (["Crude Oil Production, 48 States (Thousand Barrels per Day)","Crude Oil Production, Alaska (Thousand Barrels per Day)","Crude Oil Production, Total (Thousand Barrels per Day)","Natural Gas Liquids Production (Thousand Barrels per Day)","Renewable Fuels and Oxygenate Plant Net Production (Thousand Barrels per Day)","Petroleum Processing Gain (Thousand Barrels per Day)","Petroleum Net Imports (Thousand Barrels per Day)","Petroleum Adjustments (Thousand Barrels per Day)"],axis=1)
clean_petroleum
```

```python
# Clean out renewable data - kept all data
clean_renewables = renewables_df
clean_renewables
```

```python
# Merging data - part 1
combined1_df = pd.merge(clean_coal,clean_renewables, on="Month")
```

```python
# Merging data - part 2
combined2_df = pd.merge(clean_petroleum,clean_natural_gas, on="Month")
```

```python
# Merging data - part 3
final_combine = pd.merge(combined1_df ,combined2_df, on="Month")
final_combine
```

```python
# Dropping year 2020
drop_half = final_combine.drop(final_combine.index[564:571])
drop_half
```

```python
# Removing data before 1980
drop_final = drop_half.drop(drop_half.index[0:84])
drop_final
```
